```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['font.size'] = 12
```

```python
x = np.linspace(0, 1, 100)
fig = plt.figure(figsize=(8, 10))
gs = plt.GridSpec(2, 2, width_ratios=[1, 0.5], height_ratios=[0.3, 1])

# Cross-axis section
ax = fig.add_subplot(gs[0, 0])
ax.set_title('Cross-axis section (infinite channel)', fontsize=12)
ax.set_xlim([1, 0])
ax.set_ylim([0, 1])
ax.xaxis.set_ticks([0, 0.2, 1])
ax.yaxis.set_ticks([0, 0.7, 1])
ax.xaxis.set_ticklabels('')
ax.yaxis.set_ticklabels(['$-H$', '$-h_1$', 0])
ax.set_ylabel('$z \longrightarrow$')
ax.text(0.99, 0.87, '(a)')
ax.plot([0, 1], [0.7, 0.7], 'k--')
ax.text(0.2, 0.4, '$\\zeta_{side}(x, t)$')
ax.arrow(0.05, 0.55, 0, 0.22, head_width=0.015, head_length=0.03, fc='k', zorder=10)
for a, color in zip([0.002, 0.003, 0.004], ['lightgray', 'gray', 'k']):
    ax.plot(x, -a*np.sinh(10*(x-0.5))+0.7, '-', color=color)
ax.text(0.5, 0.85, 'Ekman flux', ha='center')
ax.arrow(0.35, 0.8, 0.3, 0, head_width=0.035, head_length=0.02, fc='k')
ax.text(0.5, 0.3, 'downwind coastal jets', ha='center')
for xpos, dx in zip([0.1, 0.9], [-0.12, 0.12]):
    ax.arrow(xpos-(1.3*dx), 0.41, dx, 0.4, head_width=0.015, head_length=0.03, fc='k', zorder=10)
    ax.text(xpos, 0.85, '$\\odot$', ha='center', fontsize=16)

# Along-axis section
ax = fig.add_subplot(gs[1, 1])
ax.set_title('Along-axis section\n(infinite channel)', fontsize=12)
ax.set_xlim([0, 1])
ax.set_ylim([1, 0])
ax.xaxis.set_ticks([0, 0.7, 1])
ax.yaxis.set_ticks([0, 0.13, 1])
ax.xaxis.set_ticklabels(['$-H$', '$-h_1$', 0])
ax.yaxis.set_ticklabels('')
ax.set_xlabel('$z \longrightarrow$')
ax.text(0.02, 0.04, '(b)')
ax.plot([0.7, 0.7], [0, 1], 'k--')
ax.plot(-0.00025*np.sinh(15*(x-0.5))+0.7, x, 'k-')
ax.text(0.25, 0.05, '$\\zeta_{end}(y)$', va='center')
ax.arrow(0.6, 0.05, 0.15, 0, head_width=0.015, head_length=0.03, fc='k', zorder=10)
ax.text(0.85, 0.52, 'Ekman flux', ha='center', va='center', rotation=-90)
ax.text(0.85, 0.41, '$\\odot$', ha='center', fontsize=16)
ax.text(1.21, 0.5, 'uniform along-axis wind stress ($\\tau$)', va='center', rotation=-90)
ax.arrow(1.2, 0.2, 0, 0.6, head_width=0.03, head_length=0.015, fc='k', clip_on=False)

# Plan view
ax = fig.add_subplot(gs[1, 0])
ax.set_title('Plan View (closed box)', fontsize=12)
ax.set_xlim([1, 0])
ax.set_ylim([1, 0])
ax.xaxis.set_ticks([0, 0.2, 1])
ax.yaxis.set_ticks([0, 0.13, 1])
ax.xaxis.set_ticklabels([0, '$L_R$', '$L_C$'])
ax.yaxis.set_ticklabels([0, '$L_R$', '$L_A$'])
ax.set_xlabel('$\longleftarrow x$')
ax.set_ylabel('$\longleftarrow y$')
ax.text(0.99, 0.04, '(c)')
for fac, xpos, label in zip([-1, 1], [0.8, 0.55], ['Divergence', 'Convergence']):
    ax.text(0.5+fac*0.3, 0.5+fac*0.43, label, ha='center', va='center', color='r', weight='bold')
    #ax.text(0.5-fac*0.15, 0.5+fac*0.47, 'wave propagation', ha='center', va='center', color='r')
    #ax.arrow(0.5+fac*0.05, 0.5+fac*0.44, -fac*0.35, 0, head_width=0.015, head_length=0.02, color='r')
    ax.text(0.5-fac*0.14, 0.5+fac*0.45, 'pressure gradient force', ha='center', va='center', color='b')
    ax.arrow(xpos, 0.5+fac*0.41, -0.35, 0, head_width=0.015, head_length=0.02, color='b')
ax.text(0.5, 0.48, 'Ekman flux', ha='center')
for xpos in [0.1, 0.9]:
    ax.text(xpos-0.01, 0.5, 'downwind coastal jet', va='center', rotation=-90)
    ax.arrow(xpos, 0.15, 0, 0.7, head_width=0.02, head_length=0.015, fc='k')
for ypos in np.arange(0.15, 1, 0.175):
    ax.arrow(0.25, ypos, 0.5, 0, head_width=0.015, head_length=0.02, fc='k')
fig.savefig('/home/bmoorema/Desktop/revision_figs/twolayermodel.pdf', bbox_inches='tight')
```
